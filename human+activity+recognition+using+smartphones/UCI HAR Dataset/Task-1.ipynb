```python


# # Assuming acc_x, acc_y, and acc_z are numpy arrays representing the acceleration in x, y, and z directions
# # And labels is a numpy array containing the class labels for each time series sample

# # 1. Compute the total acceleration
# w=np.random.randint(0,21)
# acc_total = np.sqrt((X_train[y_train==1][w,:,:]**2).sum(axis=1))

# # # 2. Standardize the data before applying PCA
# scaler = StandardScaler()
# acc_total_standardized = scaler.fit_transform(acc_total.reshape(-1, 1))

# # 3. Apply PCA to reduce the data to 2 dimensions
# pca = PCA(n_components=2)
# acc_pca = pca.fit_transform(acc_total_standardized)

# # 4. Plot the scatter plot to visualize the different classes
# plt.figure(figsize=(10, 7))
# for activity in np.unique(labels):
#     plt.scatter(acc_pca[labels == activity, 0], acc_pca[labels == activity, 1], label=f'Activity {activity}')
# plt.title('PCA of Total Acceleration')
# plt.xlabel('Principal Component 1')
# plt.ylabel('Principal Component 2')
# plt.legend()
# plt.show()
total_acceleration_train = np.sqrt((X_train ** 2).sum(axis=2))
total_acceleration_test = np.sqrt((X_test ** 2).sum(axis=2))

# Step 2: Flatten the 500 time steps to apply PCA
# Flattening shape from (126, 500) to (126, 500) is unnecessary here because PCA will handle it, 
# but keeping it as 2D for compatibility with PCA
train_data_flattened = total_acceleration_train
test_data_flattened = total_acceleration_test

# Step 3: Standardize the data
scaler = StandardScaler()
train_data_scaled = scaler.fit_transform(train_data_flattened)
test_data_scaled = scaler.transform(test_data_flattened)

# Step 4: Apply PCA to reduce dimensionality to 2 components
pca = PCA(n_components=2)
train_pca = pca.fit_transform(train_data_scaled)
test_pca = pca.transform(test_data_scaled)
print(train_pca.shape)

# Step 5: Plot the PCA result as a scatter plot
plt.figure(figsize=(10, 7))

activity_labels=['WALKING','WALKING UPSTAIRS','WALKING DOWNSTAIRS','SITTING','STANDING','LAYING']
# Visualizing different activities with different colors
for activity in np.unique(y_train):
    plt.scatter(train_pca[y_train == activity, 0], train_pca[y_train == activity, 1], label=f'Activity {activity}')
plt.title('PCA of Total Acceleration (Training Data)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(activity_labels)
plt.show()
```
